# Data Wrangling

**The column of customerID is not important so i will drop it**

```python
df.drop('customerID',axis=1,inplace=True)
df.head(2)
```

```python
len(df[df['TotalCharges']==" "])
```

**Converting the type of totalcharges column**

```python

df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df['TotalCharges']=df['TotalCharges'].fillna(df['TotalCharges'].median())
len(df[df['TotalCharges']==" "])
```

```python
df['Churn'].value_counts()
```

```python
df['Churn']=df['Churn'].apply(lambda x:1 if x=='Yes' else 0)
df['Churn'].value_counts()# imbalance 
```

```python
df['SeniorCitizen']=df['SeniorCitizen'].astype('object')
```

```python
categorical=list(df.select_dtypes(include=['object']).columns)
numerical=list(df.select_dtypes(include=['number']).columns)
```

**Calculating features importance**

```python
# calculate the feature impotance
global_mean=df['Churn'].mean()
round(global_mean,2)
```

**Gender**

```python
churn_gender=df.groupby('gender').Churn.mean()
gender_risk= churn_gender/global_mean
gender_risk 
# The differance between rates is small so gender is not important feature
```

**Partner**

```python
churn_partner=df.groupby('Partner').Churn.mean()
partner_risk= churn_partner/global_mean
partner_risk
# The differance between rates is not small so Partner is important feature
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
                       'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']

label_encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

# Splitting dataset #

```python
x=df.drop(columns=['Churn'],axis=1)
y=df['Churn']
```

```python
y.value_counts()# imbalance
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=44)

print(f"the shape of x_train is : {x_train.shape}")
print(f'the shape of x_test is : {x_test.shape}')
print(f'the shape of y_tain is : {y_train.shape}')
print(f'the shape of y_test is {y_test.shape}')
```

# Models #

**DecisionTreeClassifier**

```python
dt=DecisionTreeClassifier(criterion="entropy", max_depth = 4)
model2=dt.fit(x_train,y_train)
y_pred=model2.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
tree.plot_tree(model2)
plt.show()
```

**LogisticRegression**

```python
lr=LogisticRegression()
model3=lr.fit(x_train,y_train)
y_pred=model3.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=model3.classes_
)

plt.figure(figsize=(4, 3))
disp.plot(cmap="Blues", values_format="d")
plt.title("Logistic Regression – Confusion Matrix")
plt.show()
```

```python
import pandas as pd

results = pd.DataFrame({
    "Model": ["Decision Tree", "Logistic Regression"],
    "Accuracy": [0.78, 0.80],
    "Churn_Precision": [0.63, 0.65],
    "Churn_Recall": [0.35, 0.52],
    "Churn_F1_Score": [0.45, 0.57]
})

results
```

## Final Conclusion: Churn Prediction Models

Two models were evaluated for customer churn prediction: Decision Tree and Logistic Regression.

The Decision Tree model achieved an accuracy of 78%, but its recall for churn customers was only 35%. This indicates that the model fails to identify a large portion of actual churners, making it unreliable for churn-focused business decisions despite good performance on non-churn customers.

Logistic Regression achieved a higher accuracy of 80% and significantly improved churn recall at 52%. Its higher F1-score for the churn class shows a better balance between precision and recall, making it more effective at detecting customers who are likely to leave.

Overall, Logistic Regression performs better for churn prediction. Since identifying churn customers is more important than overall accuracy, Logistic Regression is the preferred model for this problem.
